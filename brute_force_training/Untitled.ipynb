```python
import pandas as pd
from glob import glob
import numpy as np
import matplotlib.pyplot as plt
```

```python
pass_data = glob("pass_data/*")
min_aucs, max_aucs, pass_ix = [], [], []
for ix in range(len(pass_data)):
    pass_data = pd.read_csv(f"pass_data/p{ix}/stats.csv", index_col=0)
    min_ado = min(pass_data.ado)
    max_ado = max(pass_data.ado)
    min_auc = min(pass_data.auc)
    max_auc = max(pass_data.auc)
    pass_ix.append(ix)
    min_aucs.append(min_auc)
    max_aucs.append(max_auc)
plt.plot(pass_ix, min_aucs)
plt.plot(pass_ix, max_aucs)
plt.show()
    
```
